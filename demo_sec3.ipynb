```python
import numpy as np
import scipy.stats as ss
from sklearn.metrics import mutual_info_score

import matplotlib.pyplot as plt
%matplotlib inline

from Machine_Learning_for_Asset_Managers.ch3_metrics import *
```

```python
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
bins=10 # descretize sample space
```

```python
#codesnippet 3.1
cXY = np.histogram2d(x, y, bins)[0]
hX = ss.entropy(np.histogram(x, bins)[0]) #marginal 
hY = ss.entropy(np.histogram(y, bins)[0]) #marginal
iXY = mutual_info_score(None, None, contingency=cXY)
iXYn = iXY/min(hX, hY) #normalized mutual information
hXY = hX+hY - iXY #joint
hX_Y = hXY-hY #conditional
hY_X = hXY-hX #contitional
```

```python
#codesnippet 3.4
size, seed = 5000, 0 
np.random.seed(seed)
x = np.random.normal(size=size)
e = np.random.normal(size=size)
y = 0*x+e
nmi = mutualInfo(x,y,True)
corr = np.corrcoef(x,y)[0,1]
```

```python
# figure 3.2
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.text(0.05, 0.9, f"corr={corr:.4f}\n nmi={nmi:.4f}", transform=ax.transAxes)
ax.set_title(r"$y=0x+\varepsilon$")
plt.show()
```

```python
# figure 3.3
size, seed = 5000, 0 
np.random.seed(seed)
x = np.random.normal(size=size)
e = np.random.normal(size=size)
y = 100 * x + e
nmi = mutualInfo(x,y,True)
corr = np.corrcoef(x,y)[0,1]

fig, ax = plt.subplots()
ax.scatter(x, y)
ax.text(0.05, 0.9, f"corr={corr:.4f}\n nmi={nmi:.4f}", transform=ax.transAxes)
ax.set_title(r"$y=100x+\varepsilon$")
plt.show()
```

```python
# figure 3.4
size, seed = 5000, 0 
np.random.seed(seed)
x = np.random.normal(size=size)
e = np.random.normal(size=size)
y = 100 * np.abs(x) + e
nmi = mutualInfo(x,y,True)
corr = np.corrcoef(x,y)[0,1]

fig, ax = plt.subplots()
ax.scatter(x, y)
ax.text(0.1, 0.9, f"corr={corr:.4f}\n nmi={nmi:.4f}", transform=ax.transAxes)
ax.set_title(r"$y=100|x|+\varepsilon$")
plt.show()
```
